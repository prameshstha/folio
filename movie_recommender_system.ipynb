```python
import ast
```

```python
def convert_to_list(movie_object):
  name_list = []
  for obj in ast.literal_eval(movie_object):
    name_list.append(obj['name'])
  return name_list
```

```python
final_movies['genres'] = final_movies['genres'].apply(convert_to_list)
```

```python
final_movies['keywords'].apply(convert_to_list)
```

```python
final_movies['keywords'] = final_movies['keywords'].apply(convert_to_list)
```

```python
def cast_to_list(cast_object):
  cast_list = []
  counter = 0
  for obj in ast.literal_eval(cast_object):
    if counter !=3:
      cast_list.append(obj['name'])
      counter += 1
    else:
      break
  return cast_list
```

```python
final_movies['cast'] = final_movies['cast'].apply(cast_to_list)
```

```python
def fetch_director(crew):
  name_list = []
  for obj in ast.literal_eval(crew):
    if obj['job'] == 'Director':
      name_list.append(obj['name'])
      break
  return name_list
```

```python
final_movies['crew'] = final_movies['crew'].apply(fetch_director)
```

```python
final_movies['overview']  = final_movies.overview.apply(lambda x:x.split())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# to stem the word, to get the simmilar words cut off such ['loving', 'loved', 'lovely', 'love'] will change to ['love', 'love', 'love', 'love']
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
ps.stem('loving')
```

```python
def stem_words(word):
  y = []
  for i in word.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
stem_words('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron')
```

```python
new_df.tags = new_df.tags.apply(stem_words)
```

```python
cv =  CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df.tags).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
## find cosine distance between two vector movies
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vector_similarity = cosine_similarity(vectors)
```

```python
vector_similarity[0]
```

```python
## enumerate will give index and value in 'vector_similarity'
list(enumerate(vector_similarity[0]))
## sort the enumerate vector_similarity in decending order with the vector_similarity value, thatis why in lambda it is x[1]
sorted(list(enumerate(vector_similarity[0])), reverse=True, key=lambda x:x[1])
```

```python
## [1:6] to get the top 5 similar movie from vector_similarity[n] as it is sorted already
sorted(list(enumerate(vector_similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
sorted(vector_similarity[0], reverse=True)
```

```python
def recommend_by_title(movie):
  #get index of the given movie
  movie_index = new_df[new_df.title == movie].index[0]
  # get the vector_similarty with the same index with movie_index as it has same index
  distances = vector_similarity[movie_index]
  recommended_movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
  for i in recommended_movie_list:
    print(new_df.iloc[i[0]].title)
```

```python
def recommend_by_id(movie_id):
  #get index of the given movie
  movie_index = new_df[new_df.id == movie_id].index[0]
  # get the vector_similarty with the same index with movie_index as it has same index
  distances = vector_similarity[movie_index]
  recommended_movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
  for i in recommended_movie_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend_by_title('Avatar')
```
